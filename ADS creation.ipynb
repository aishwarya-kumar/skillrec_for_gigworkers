```python
import pandas as pd
```

```python
csv_list = ['worker_profiles_DA', 'worker_profiles_DS','worker_profiles_web', 'worker_profiles_GD', 'worker_profiles_MKT']
```

```python
dataframes = []
for file_name in csv_list:
    file_path =f"{file_name}.csv"
    df = pd.read_csv(file_path)
    print(f"Shape of {file_name}: {df.shape}")
    dataframes.append(df)
```

```python
dataframes[0]
```

```python
worker_profiles = pd.concat(dataframes, ignore_index=True)
worker_profiles = worker_profiles.loc[:, ~worker_profiles.columns.str.contains('^Unnamed')]
worker_profiles.shape
```

```python
worker_profiles.head()
```

```python
print("Missing values:\n", worker_profiles.isnull().sum())
```

```python
worker_profiles['country'] = worker_profiles['country'].fillna('')
```

```python
print("Missing values:\n", worker_profiles.isnull().sum())
```

```python
worker_profiles.shape
```

```python
worker_profiles_vf= worker_profiles.dropna()
```

```python
worker_profiles_vf = worker_profiles_vf.sample(frac = 1)
```

```python
worker_profiles_vf.reset_index(drop=True, inplace=True)
```

```python
worker_profiles_vf.to_csv("worker_profiles_vf.csv")
```
